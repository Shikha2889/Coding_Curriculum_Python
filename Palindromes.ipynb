# 1.Check whether a given string is a palindrome or not 






```python
def check_palindrome(string):
    start = 0
    end = len(string)-1
    palindrome= True
    while start<len(string)//2:
        if string[start]!=string[end]:
            palindrome = False
        start+=1
        end-=1
    return palindrome

check_palindrome('aba')
```

#2.Program to check if a given number is a palindrome or not
#For example, 12321 is palindrome, but 1451 is not palindrome. 
image.png

```python
num = int(input("Enter any number"))
def reverse_number(num):
    rev = 0
    while num>0:
        digit = num%10
        rev = rev*10+digit
        num = num//10
    return rev


# Time Complexity --O(k)

#Check whether a given number is palindrome or not 
def isPalindrome(num):
    rev = reverse_number(num)
    if (rev == num):
        return True
    else:
        return False



```

#3.Program to print all palindromes in a given range 
# For example if the given range is {10, 115}, then output should be {11, 22, 33, 44, 55, 66, 77, 88, 99, 101, 111}

#Hint :We can run a loop from min to max and check every number for palindrome. If the number is a palindrome, we can simply print it. 

```python
def palindrome_range(A,B):
    for i in range(A,B+1):
        
        if(isPalindrome(i)):
            print(i,end =' ')

palindrome_range(10,115)
        
```

#6.Sentence Palindrome
#Sentence Palindrome (Palindrome after removing spaces, dots, .. etc)
#Write a program to check if a sentence is a palindrome or not. You can ignore white spaces and other characters to consider sentences as palindrome.
#Examples: 

#Input : str = "Too hot to hoot."
#Output : Sentence is palindrome.

#Input : str = "Abc def ghi jklm."
#Output : Sentence is not palindrome.

```python
def sentence_palindrome(text):
    new_text=''
    start = 0
    end = len(new_text)-1
    palindrome = True
    for i in text:
        if(i.isalnum()):
            new_text+=i
    while start<end:
        if(new_text[start]!=new_text[end]):
            palindrome = False
        start+=1
        end-=1
    if(palindrome == False):
        print("Sentence is not Palindrome")
    else:
        print("Sentence is palindrome")
sentence_palindrome("Too hot to hoot")
```
